```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
```

```python
data = pd.read_csv('C:/Users/Seo/Downloads/ALL UFC FIGHTERS 2_23_2016 SHERDOG.COM - Sheet1.csv',encoding= 'unicode_escape')
data.head()
# ufc 파이터들 목록
```

```python
g = nx.Graph()
g = nx.nx.from_pandas_edgelist(data, source = 'name', target = 'country')
print(nx.info(g))
```

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(g, k = 0.15)
nx.draw_networkx(g,pos, node_size = 25, node_color = 'blue')
plt.show()
# 나라를 기준으로 연관성 분석
```
